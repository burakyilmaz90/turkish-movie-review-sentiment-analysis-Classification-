## Problem Definition

The goal of this project is to develop a machine learning model that predicts whether movie user reviews are 
positive or negative.

This problem has been addressed in the field of natural language processing (NLP) 
under sentiment analysis and 
has been modeled as a binary classification problem.


# 1. Data Collection

In this section, the data to be used in the project was collected via the web. The goal is to obtain the raw data set to be used in the analysis.


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

filmler = {
    "7. Kogustaki Mucize": "275065",
    "Babam ve Oğlum": "111471",
    "Veda": "180472",
    "Ayla": "249882",
    "The Dark Knight": "132874",
    "Dünyayı Kurtaran Adamın Oğlu": "125161",
    "Superman": "229669",
    "Dag 2": "247803",
    "Harry Potter ve Ölüm Yadigarları" :"126693",
    "Matrix Resurrections" :"254560",
    "The Godfather" :"1628",
    "Black" : "60969",
    "Recep İvedik 6" : "258775",
    "Cumali Ceber" : "256949"
}
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

tum_veriler = []
bakilacak_sayfa_sayisi = 12
print(f"İşlem başladı")
for film_adi, film_id in filmler.items():
    print(f"\n{film_adi} taranıyor")
    
    for sayfa_no in range(1, bakilacak_sayfa_sayisi + 1):
        url = f"https://www.beyazperde.com/filmler/film-{film_id}/kullanici-elestirileri/?page={sayfa_no}"
        print(f"Sayfa {sayfa_no} okunuyor...")
        
        try:
            response = requests.get(url, headers=headers)
            if response.status_code != 200:
                print(f"Hata: Sayfa {sayfa_no} alınamadı (Kod: {response.status_code})")
                break
    
            soup = BeautifulSoup(response.content, 'html.parser')        
            
            ana_bolum = soup.select_one('#content-layout > section > div > section:nth-of-type(2)')
            
            if not ana_bolum:
                print("Yorum alanı bulunamadı, bir sonraki filme geçiliyor.")
                break
                
            yorum_kartlari = ana_bolum.find_all("div", class_="review-card")
            
            if not yorum_kartlari:
                print("Bu sayfada yorum yok, sonraki filme geçiliyor.")
                break
                
            for kart in yorum_kartlari:
                puan_etiketi = kart.find("span", class_="stareval-note")
                puan = puan_etiketi.text.strip() if puan_etiketi else "N/A"
                
                yorum_etiketi = kart.find("div", class_="content-txt")
                yorum = yorum_etiketi.text.strip() if yorum_etiketi else "Yorum yok"
                
                tum_veriler.append({
                    "Puan": puan,
                    "Yorum": yorum
                })
            time.sleep(1)
        except Exception as e:
            print(f"Bir hata oluştu: {e}")
            break

df = pd.DataFrame(tum_veriler)

# --- Kontroller ---
print("Toplam satır:", len(df))
print("Kolonlar:", df.columns.tolist())
print(df.head(3))

# Boş yorumları temizle 
df["Yorum"] = df["Yorum"].astype(str).str.strip()
df = df[df["Yorum"].str.len() > 0].copy()

# --- Kaydet ---
df.to_csv("raw_reviews.csv", index=False, encoding="utf-8-sig")
print(f"Kaydedildi raw_reviews.csv | Satır sayısı: {len(df)}")
```
